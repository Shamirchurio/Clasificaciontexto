```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
O1={'Name':['m1','m2','m3','m4'],
    '0':[65.81,67.47,76.51,62.87],
    '1':[63.9,42.76,0,46.6],
    '2 ':[90.76,94.54,93.36,76.28],
    'test_loss':[42.76,58.58,63.72,65.43],
    'test_accuracy':[88.16,84.21,76.32,80.26],
    'val_loss':[51.06,63.71,70.93,80.81],
    'val_accuracy':[84.00,82.67,81.33,78.67]
    }
```

```python
df=pd.DataFrame.from_dict(O1);df
```

```python
ypoints = df[['0','1','2 ']]
ypoints.index=['m1','m2','m3','m4']

ypoints1 = df[['test_accuracy','val_accuracy']]
ypoints1.index=['m1','m2','m3','m4']

ypoints2 = df[['test_loss','val_loss']]
ypoints2.index=['m1','m2','m3','m4']

###
plt.plot(ypoints, linestyle = 'dotted', marker = 'o')
plt.gca().legend(('0','1','2'))
plt.title('Promedio de acertividad etiquetas')
plt.show()

plt.plot( ypoints1, linestyle = 'dotted', marker = 'o')
plt.gca().legend(('test_accuracy','val_accuracy'))
plt.title('Accuracy en los datos de validación y test')
plt.show()

plt.plot( ypoints2, linestyle = 'dotted', marker = 'o')
plt.gca().legend(('test_loss','val_loss'))
plt.title('Loss en los datos de validación y test')
plt.show()
```

```python
O2={'Name':['Análisis de calidad baja','Análisis de calidad media','Análisis de calidad'],
    'Análisis de calidad baja':[65.81,9.81,24.37],
    'Análisis de calidad media':[17.55,63.9,18.54],
    'Análisis de calidad':[7.7,1.54,90.76]
    }
```

```python
O2
```

```python
df1=pd.DataFrame.from_dict(O2);df1
```

```python
ypoints = df1[['Análisis de calidad baja',
  'Análisis de calidad media',
  'Análisis de calidad']]

ypoints.index=['Análisis de calidad baja',
  'Análisis de calidad media',
  'Análisis de calidad']

plt.plot(ypoints, linestyle = 'dotted', marker = 'o')
plt.gca().legend(('Análisis de calidad baja',
  'Análisis de calidad media',
  'Análisis de calidad'))
plt.title('Promedio de acertividad etiquetas')
plt.show()
```

```python
eje_x=['Análisis de calidad media','Análisis de calidad','Análisis de calidad baja']
filtrada2 = [65.81,9.81,24.37]
```

```python
fig, ax = plt.subplots(1, 1)
ax.bar(eje_x,filtrada2, color='green' )

# Set title
ax.set_title("Porcentaje predicción etiqueta análisis de calidad media")
  
# adding labels
ax.set_xlabel('Etiquetas')
ax.set_ylabel('Porcentaje')
  
# Make some labels.
rects = ax.patches
labels = [round(i*1,2) for i in 
    filtrada2
]
  
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width() / 2, height+0.01, label,
            ha='center', va='bottom')
  
# Show plot
plt.xticks(rotation=45)
plt.show()
  
```
